1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even_numbers(numbers):
    sum_even = 0
    for num in numbers:
        if num % 2 == 0:
            sum_even += num
    return sum_even
```

```python
sum_even_numbers([2,5,89,74,56,22,11,568])
```

 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms

```python
class FibonacciIterator:
  def __init__(self, max_terms):
    self.max_terms = max_terms
    self.current_term = 0
    self.a = 0
    self.b = 1

  def __iter__(self):
    return self

  def __next__(self):
    if self.current_term >= self.max_terms:
      raise StopIteration

    result = self.a
    self.a, self.b = self.b, self.a + self.b
    self.current_term += 1
    return result
```

 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
  for i in range(exponent + 1):
    yield 2 ** i

powers_generator = powers_of_two(5)
for power in powers_generator:
  print(power)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
file_content = """This is the first line.
This is the second line.
This is the third line."""

with open("sample.txt", "w") as f:
    f.write(file_content)
```

```python
def read_lines_generator(file_path):
  with open(file_path, 'r') as f:
    for line in f:
      yield line
```

```python
lines_generator = read_lines_generator("sample.txt")
for line in lines_generator:
  print(line.strip())
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius_temperatures = [0.0, 10.0, 20.0, 30.0, 40.0]
def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32

fahrenheit_temperatures_map = map(celsius_to_fahrenheit, celsius_temperatures)
fahrenheit_temperatures = list(fahrenheit_temperatures_map)
print(fahrenheit_temperatures)
```

11) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
 Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

```python
def calculate_order_totals(orders):

    result = []
    for order in orders:
        order_number = order[0]
        quantity = order[2]
        price_per_item = order[3]

        total_price = quantity * price_per_item

        if total_price < 100.00:
            total_price += 10.00

        result.append((order_number, total_price))
    return result

book_orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

calculated_totals = calculate_order_totals(book_orders)
print(calculated_totals)
```

**5. What are iterators in Python and how do they differ from iterables?**

--Iterables
An iterable is an object that can be looped over, meaning you can traverse through all its elements one by one. Think of an iterable as a container of data, such as a list, tuple, string, or even a dictionary (iterating over its keys by default).
To be considered iterable, an object needs to implement either:
The special method __iter__(), which should return an iterator object for the iterable.
The __getitem__() method, especially for sequence types

Iterators

An iterator is an object that specifically facilitates the process of iterating over an iterable object. It acts as a "pointer" or "bookmark" within the sequence, keeping track of its current position and knowing how to access the next element.
Iterators are defined by implementing the iterator protocol, which consists of two methods:
__iter__(): This method returns the iterator object itself.
__next__(): This method retrieves the next element from the iterator and raises a StopIteration exception when there are no more elements to return.
The built-in iter() function can be used to obtain an iterator from an iterable object, while the next() function is used to retrieve the next element from an iterator.

 6. Explain the concept of generators in Python and how they are defined.

In Python, generators are special functions that produce a sequence of values on demand, rather than creating the entire sequence in memory at once. This makes them ideal for working with large datasets or infinite sequences, as they are memory-efficient and allow for lazy evaluation.

Generators can be defined in two ways:

Generator functions: These are defined using def but use the yield keyword instead of return to produce values and pause execution. An example is provided in the source.

Generator expressions: These are a concise way to create generators, similar to list comprehensions but using parentheses.

The yield keyword is central to generator functions. When used, it pauses the function, returns a value, and saves the function's state, allowing it to resume later.


def count_up_to(n):

i = 0

while i < n:

yield i

i += 1

# Using the generator
my_generator = count_up_to(5)


print(next(my_generator))  # Output: 0

print(next(my_generator))  # Output: 1


for num in my_generator:

print(num)  # Output: 2, 3, 4


7. What are the advantages of using generators over regular functions?

Generators offer several significant advantages over regular functions, especially when dealing with large datasets or streams of data. Here are the main benefits:

1. Memory Efficiency

This is the most significant advantage. A regular function that returns a collection (like a list) creates the entire collection in memory at once. If the collection is very large, this can consume a lot of memory and even crash your program.

A generator, on the other hand, yields one item at a time. It doesn't store the entire sequence in memory. It only keeps track of its current state, which requires very little memory.

Example: Generating a large sequence of numbers

Let's compare a function that returns a list of a million numbers with a generator that does the same:

import sys

# Regular function
def first_n_numbers_list(n):

  nums = []

  num = 0

  while num < n:

    nums.append(num)

    num += 1

  return nums


# Generator function
def first_n_numbers_generator(n):

  num = 0

  while num < n:

    yield num

    num += 1


# Compare memory usage
list_of_numbers = first_n_numbers_list(1000000)

print(f"Memory usage of list: {sys.getsizeof(list_of_numbers)} bytes")


generator_of_numbers = first_n_numbers_generator(1000000)

print(f"Memory usage of generator: {sys.getsizeof(generator_of_numbers)} bytes")

When you run this, you'll see that the list takes up a significant amount of memory, while the generator object is very small, regardless of the number of items it can produce.

2. Ability to Represent Infinite Sequences

Because generators produce values one at a time and don't store the whole sequence, they can be used to represent infinite sequences of data. This is impossible with a regular function that returns a list.

Example: An infinite sequence of natural numbers


def all_natural_numbers():

  num = 1

  while True:

yield num

num += 1



natural_numbers = all_natural_numbers()

print(next(natural_numbers)) # Output: 1

print(next(natural_numbers)) # Output: 2

print(next(natural_numbers)) # Output: 3

# and so on...

3. Lazy Evaluation


Generators use lazy evaluation, which means they compute values only when they are requested. This can lead to performance improvements if you don't need to use all the values in the sequence. For example, if you are searching for a specific item in a large dataset, a generator will stop producing values as soon as the item is found.
